```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_sides = 8
```

Data from the last 5 matches (start from latest):

```python
Goals_per_game = [4, 4, 5, 3, 1]
Fouls_per_game = [9, 11, 14, 9, 9]
Pass_completion_per = [92.3, 76.6, 90.0, 86.6, 85.3]
Cross_completion_per = [31.2, 27.3, 41.7, 37.5, 12.5]
Assists_per_game = [3, 2, 3, 3, 1]
Shots_on_target_ratio = [0.25, 0.47, 0.5, 0.42, 0.3]
Shots_per_game = [20, 15, 20, 12, 10]
Non_penalty_xG_90 = [4, 3, 5, 3, 1]
```

Average the data:

```python
GPG = sum(Goals_per_game) / len(Goals_per_game)
FPG = sum(Fouls_per_game) / len(Fouls_per_game)
PCR = sum(Pass_completion_per) / len(Pass_completion_per)
CCR = sum(Cross_completion_per) / len(Cross_completion_per)
APG = sum(Assists_per_game) / len(Assists_per_game)
SOTR = sum(Shots_on_target_ratio) / len(Shots_on_target_ratio)
SPG = sum(Shots_per_game) / len(Shots_per_game)
NPXG = sum(Non_penalty_xG_90) / len(Non_penalty_xG_90)
```

```python
data = [GPG, FPG, PCR, CCR, APG, SOTR, SPG, NPXG]
print(data)
```

Normalise data: 

```python
max_values = [5.0, 25.0, 100.0, 100.0, 5.0, 1.0, 25.0, 5.0]
normalized_data = [d / m for d, m in zip(data, max_values)]

normalized_data += normalized_data[:1]
```

```python
labels = [
    f'Goals per game: {data[0]:.2f}',
    f'Fouls per game: {data[1]:.2f}',
    f'Pass completion %: {data[2]:.2f}',
    f'Cross completion %: {data[3]:.2f}',
    f'Assists per game: {data[4]:.2f}',
    f'Shots on target ratio: {data[5]:.2f}',
    f'Shots per game: {data[6]:.2f}',
    f'Non-penalty xG/90: {data[7]:.2f}'
]
```

```python
# Compute the angle of each axis
angles = np.linspace(0, 2 * np.pi, n_sides, endpoint=False).tolist()
angles += angles[:1]  # Close the loop

# Create the plot
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

# Plot the data
ax.plot(angles, normalized_data, linewidth=2, linestyle='solid')
ax.fill(angles, normalized_data, 'skyblue', alpha=0.4)

# Set the labels with data values
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, size=10)

# Hide the radial axis labels (the floats inside the chart)
ax.yaxis.set_visible(False)

# Add title and grid
plt.title('Team Attack', size=15)
ax.grid(True)

plt.show()
```
